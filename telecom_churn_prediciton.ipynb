```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('../input/customerattritionprediction/train.csv')
test = pd.read_csv('../input/customerattritionprediction/test.csv')
sample_submission = pd.read_csv('../input/customerattritionprediction/Sample Submission.csv')
```

```python
train
```

```python
train.loc[pd.isna(train['GrandPayment']), :].index
```

```python
test.loc[pd.isna(test['GrandPayment']), :].index
```

```python
#train['GrandPayment'].fillna(2290.544,inplace=True)
#test['GrandPayment'].fillna(2290.544,inplace=True)
for i in [643, 1169, 1186, 2281, 3611, 3918, 5579, 5742, 5796, 6203]:
    train['GrandPayment'].iloc[i] = train['ServiceSpan'].iloc[i]*train['QuarterlyPayment'].iloc[i] 
    
test['GrandPayment'].iloc[532] = test['ServiceSpan'].iloc[532]*test['QuarterlyPayment'].iloc[532]
```

```python
train['ServiceSpan'].unique()
```

```python
for i in range(len(train)):
    if(train['ServiceSpan'].iloc[i]==-1 or train['ServiceSpan'].iloc[i]==-2):
        train['ServiceSpan'].iloc[i] = int(train['GrandPayment'].iloc[i]/train['QuarterlyPayment'].iloc[i])
```

```python
train.drop(['ID' ,'CyberProtection', 'HardwareSupport', 'TotalDependents','SettlementProcess'], axis=1, inplace=True)
test.drop(['ID', 'CyberProtection', 'HardwareSupport', 'TotalDependents','SettlementProcess'], axis=1, inplace=True)
```

```python
train.head(1)
```

```python
a=[]
b=[]
```

```python
for i in range(len(train)):
    if train['Aged'].iloc[i]=='Yes' and train['Married'].iloc[i]=='Yes':
        a.append(1)
    else:
        a.append(0)
        
for i in range(len(test)):
    if test['Aged'].iloc[i]=='Yes' and test['Married'].iloc[i]=='Yes':
        b.append(1)
    else:
        b.append(0)
            
```

```python
train['Aged_Married'] = a
test['Aged_Married'] = b
```

```python
train = pd.get_dummies(train, columns=['sex', 'Aged', 'Married', 'MobileService', '4GService', 'TechnicalAssistance', 'FilmSubscription'])
```

```python
test = pd.get_dummies(test, columns=['sex', 'Aged', 'Married', 'MobileService', '4GService', 'TechnicalAssistance', 'FilmSubscription'])
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
```

```python
train['CustomerAttrition']= label_encoder.fit_transform(train['CustomerAttrition'])
```

```python
train.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_res, X_test, y_res, y_test = train_test_split(train.drop('CustomerAttrition',axis=1), train['CustomerAttrition'], test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import ADASYN
sm = ADASYN(random_state = 2)
X_train, y_train = sm.fit_resample(X_res, y_res)
```

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
y_test = np.array(y_test)
```

```python
y_pred2 = []
for i in range(len(y_pred)):
    if y_pred[i]<0.5:
        y_pred2.append('No')
    else:
        y_pred2.append('Yes')
y_test2 = []
for i in range(len(y_test)):
    if y_test[i]<0.5:
        y_test2.append('No')
    else:
        y_test2.append('Yes')        
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test2, y_pred2)
```

```python
prediction = reg.predict(test)
```

```python
prediction2 = []
for i in range(len(prediction)):
    if prediction[i]>0.5:
        prediction2.append('No')
    else:
        prediction2.append('Yes')
```

```python
sub = sample_submission.copy()
sub['CustomerAttrition'] = prediction2
sub
sub.to_csv('submission_new2.csv', index=False)
```

```python
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
import xgboost as xgb
from catboost import CatBoostClassifier
```

```python

hyper_params1 = { "n_estimators": 60,'n_jobs':-1, 'learning_rate':0.5}

hyper_params2 = {'n_estimators': 60,'n_jobs':-1}

hyper_params3 = {'n_estimators':60, 'n_jobs':-1,'learning_rate':0.5 }

hyper_params4 = {'iterations':60,'learning_rate':0.5}
```

```python
model1 = lgb.LGBMClassifier(**hyper_params1)
model2 = RandomForestClassifier(**hyper_params2)
model3 = xgb.XGBClassifier(**hyper_params3)
model4 = CatBoostClassifier(**hyper_params4)
```

```python
estimator = []
estimator.append(('lgb', model1))
estimator.append(('rf',  model2)) 
estimator.append(('xgb', model3))
estimator.append(('cbr', model4)) 

vot = VotingClassifier(estimators = estimator) 
vot.fit(X_train, y_train)
```

```python
y_pred = vot.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
prediction = vot.predict(test)
pred = label_encoder.inverse_transform(prediction)
sub = sample_submission.copy()
sub['CustomerAttrition'] = pred
sub
sub.to_csv('submission_new4.csv', index=False)
```
